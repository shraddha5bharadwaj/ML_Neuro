```python
import pynwb
from pynwb import NWBHDF5IO

# Path to the NWB file
# file_path = 'path_to_your_file.nwb'
file_path = '/content/gdrive/MyDrive/MLNBD/Project/sub-Jenkins_ses-small_desc-train_behavior+ecephys (3).nwb'

# Open the NWB file in read mode
with NWBHDF5IO(file_path, 'r') as io:
    # Read the NWB file into a NWBFile object
    nwb_file = io.read()

    # Access general metadata
    print("Experiment Description:", nwb_file.experiment_description)
    print("Experimenters:", nwb_file.experimenter)
    print("Session Start Time:", nwb_file.session_start_time)
    print("Session Description:", nwb_file.session_description)

    # Access devices
    devices = nwb_file.devices
    for device_name, device in devices.items():
        print(f"Device: {device_name}, Type: {device.__class__.__name__}")

    # Access electrode groups
    electrode_groups = nwb_file.electrode_groups
    for group_name, group in electrode_groups.items():
        print(f"Electrode Group: {group_name}, Type: {group.__class__.__name__}")

    # Access electrodes (DynamicTable)
    electrodes = nwb_file.electrodes
    print(f"Number of electrodes: {len(electrodes)}")

    # Access trials (TimeIntervals)
    trials = nwb_file.intervals['trials']
    print(f"Number of trials: {len(trials)}")

    # Access units (Units)
    units = nwb_file.units
    print(f"Number of units: {len(units)}")

    # Access subject info
    subject = nwb_file.subject
    print(f"Subject ID: {subject.subject_id}, Species: {subject.species}, Sex: {subject.sex}")

    # Access related publications
    publications = nwb_file.related_publications
    print("Related Publications:", publications)
```
